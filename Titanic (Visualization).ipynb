 # Continuos vs categorical

```python
sb.set_style('whitegrid')
sb.boxplot(train['Pclass'],train["Age"],palette='Blues')
plt.title('Age distribution/Pclass')
plt.show()
```

older people (also more rich) travelled from class 1

```python
sb.set_style('darkgrid')
fig=sb.boxplot(train['Survived'],train['Age'],palette='Reds')
plt.title('Age distribution/Survived')
fig.set(xticklabels=["0-No","1-Yes"])
plt.show()
```

slightly more priority should be given to rescue older people

```python
sb.set_style('darkgrid')
sb.boxplot(train['Survived'],train['Fare'],palette='Blues')
plt.show()
```

people who paid more for ticket (obviously those who are travwlling from class 1/2 may had been rescued first.

```python
plt.figure(figsize=(8,4))
sb.set_style('dark')
sb.violinplot(train['Age'],train['Sex'], hue='Survived', data=train, split=True)
```

1. survival rate of male of age between 20 to 30 is vey high.
2. survival rate of females is roughly irrespective of their age.still most of the female of age 15 to 35 survived

```python
plt.figure(figsize=(8,7))
sb.set_style('white')
sb.violinplot(train['Age'],train['Title'],hue="Survived",data=train,split=True ,color='Blue')
plt.title('Age and Title vs Survived')
plt.show()
```
